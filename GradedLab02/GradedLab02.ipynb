##### __(b)  Insert the documents from the file 'restaurants.json' into the collection restaurants.__ 

```python
# Type your code here or use a MongoDB GUI.

from pymongo import MongoClient
import json

# Verbindung zu MongoDB herstellen
client = MongoClient("mongodb://localhost:27017/")  # Lokale MongoDB-Verbindung
db = client['data']  # Datenbank "data"  auswählen
collection = db['restaurants']  # Sammlung "restaurants" auswählen

# JSON-Daten aus der Datei laden und in die Sammlung einfügen
try:
    with open('restaurants_mongodb.json', 'r') as file:  
        data = json.load(file)  # JSON-Daten laden
        if isinstance(data, list):  # Prüfen, ob die Datei ein Array von Objekten ist
            collection.insert_many(data)  # Mehrere Dokumente einfügen
        else:
            collection.insert_one(data)  # Ein einzelnes Dokument einfügen
    print("Dokumente erfolgreich in die Sammlung 'restaurants' eingefügt!")
except Exception as e:
    print("Fehler beim Einfügen der Dokumente:", e)
```

##### (c)  Write the queries to print the requested results.

* (1) Write the query to print the total number of documents in the collection 'restaurants'
* (2) Write the query to print 5 restaurant names with 'Bronx' as borough from the collection 'restaurants'
* (3) Write the query to print the restaurant names that achieved a score, more than 80 but less than 100 from the c collection 'restaurants'
* (4) Write the query to print the restaurant names for those restaurants which contain 'Wil' as first three letters in its name.
* (5) Write the query to print the restaurant names which are not belonging to the borough Staten Island or Queens or Bronx or Brooklyn.
* (6) Write the query to print each borough and the number of restaurants for each borough (hint: use the grouping function)
* (7) Write the query to print the restaurant names which have the term 'Mexican' in the name attribute but don't have mexican as content in the cuisine attribute.
* (8) Write the query to print all names of bakeries with 'sugar' in the name.

```python
#1 print the total number of documents in the collection 'restaurants'
print(collection.count_documents({}))

```

```python
#2 print 5 restaurant names with 'Bronx' as borough from the collection 'restaurants'
cursor = collection.find({"borough": "Bronx"}, {"name": 1}).limit(5)
for document in cursor:
    print(document)
    
```

```python
#3 print the restaurant names that achieved a score, more than 80 but less than 100 from the c collection 'restaurants'
cursor = collection.find({"grades.score": {"$gt": 80, "$lt": 100}}, {"name": 1})
for document in cursor:
    print(document)
    
```

```python
#4 print the restaurant names for those restaurants which contain 'Wil' as first three letters in its name.
cursor = collection.find({"name": {"$regex": "^Wil"}}, {"name": 1})
for document in cursor:
    print(document)
    
```

```python
#5 print the restaurant names which are not belonging to the borough Staten Island or Queens or Bronx or Brooklyn.
cursor = collection.find({"borough": {"$nin": ["Staten Island", "Queens", "Bronx", "Brooklyn"]}}, {"name": 1})
for document in cursor:
    print(document)
```

```python
#6 print each borough and the number of restaurants for each borough (hint: use the grouping function)
# Aggregation: Gruppierung nach Borough und Zählen der Restaurants
cursor = collection.aggregate([
    {"$match": {"borough": {"$ne": None, "$ne": ""}}},  # Ausschluss von null oder leeren Boroughs
    {"$group": {"_id": "$borough", "count": {"$sum": 1}}},
    {"$sort": {"_id": 1}}  # Optional: Sortierung nach Borough-Namen
])

# Ausgabe
for document in cursor:
    print(f"Borough: {document['_id']}, Count: {document['count']}")

    
```

```python
# 7 print the restaurant names which have the term 'Mexican' in the name attribute but don't have mexican as content in the cuisine attribute
cursor = collection.find({
    "name": {"$regex": "Mexican"},
    "cuisine": {"$not": {"$regex": "mexican", "$options": "i"}}
}, {"name": 1})

for document in cursor:
    print(document)
```

```python
#8 print all names of bakeries with 'sugar' in the name.
# Abfrage der Bäckereien mit 'sugar' im Namen
cursor = collection.find({
    "name": {"$regex": "sugar", "$options": "i"},
    "cuisine": "Bakery"
}, {"name": 1})

# Verwenden Sie eine Menge, um doppelte Namen zu vermeiden
unique_names = set()

# Iteration durch die Ergebnisse und Hinzufügen zu der Menge
for document in cursor:
    unique_names.add(document["name"])

# Ausgabe der eindeutigen Namen
for name in unique_names:
    print(name)
```
